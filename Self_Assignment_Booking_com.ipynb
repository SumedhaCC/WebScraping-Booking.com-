```python
def transform(soup):
    rating =[]    
    for block in soup.find_all('div', class_="a1b3f50dcd f7c6687c3d bdf0df2d01 f920833fe5"):
        ratings = block.find('div', {'class' : 'b5cd09854e d10a6220b4'})
        if ratings:
            rating.append(ratings.get_text().strip())
        else:
            rating.append("Ratings not available")
    return rating
```

```python
def transform1(soup):
    review = []
    for block in soup.find_all('div', class_="a1b3f50dcd f7c6687c3d bdf0df2d01 f920833fe5"):
        reviews = block.find("div",attrs={'class':'d8eab2cf7f c90c0a70d3 db63693c62'})      
        if reviews:
            review.append(reviews.get_text().strip())
        else:
            review.append("Reviews not available")
    return review
```

```python
def blocks(soup):
    titles=[]
    locations = []
    from_centers = []

    for block in soup.findAll("div", {"class": "a1b3f50dcd b2fe1a41c3 a7c67ebfe5 d19ba76520 d14b211b4f"}):
        title = block.find("div",attrs={'class':'fcab3ed991 a23c043802'})
    
        location = block.find("span",attrs={'class':'f4bd0794db b4273d69aa'})
        from_center = block.find("span",attrs={'data-testid':'distance'})
        rating=transform(soup)
        review=transform1(soup)
        titles.append(title.get_text().strip())
        locations.append(location.get_text().strip())
        from_centers.append(from_center.get_text().strip())
    return titles,locations,from_centers,rating,review
```
